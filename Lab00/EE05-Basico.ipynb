#### Duck typing   en los lenguajes de programación, es una técnica para determinar si un objeto es una instacia de una determinada clase

````python
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError: # not iterable
        return False
````

````python
isiterable('a string')
isiterable([1, 2, 3])
isiterable(5)
````
