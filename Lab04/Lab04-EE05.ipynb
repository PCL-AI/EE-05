# Laboratorio 04: Proceso de Markov

Un jugador comienza con **i** soles. Cada vez, apuesta S/. 1 en un juego justo, hasta que tenga 0 o **n** soles.  Calcular: 

1. ¿Cuál es la probabilidad de que el termine con n soles? 
2. ¿Dinero (en soles) esperado al final?
3. ¿Tiempo esperado el jugador permanecer en el juego?



```
# Digamos que tenemos n soles
# Cada lanzamiento de moneda, apostamos un sol.
# ¿Cuantos lanzamientos hasta que perdamos todo?

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```
# ¿Cuanto puedes acumular antes de perder todo?

def apostador(Dinero, g,p):
    Dinero, contar = Dinero,0
    max_Dinero = Dinero

    while Dinero:
        lanzar = np.random.random()
        if lanzar < p:
            Dinero -= 1
        if lanzar > g:
            Dinero += 1
        contar += 1
   
        if Dinero > max_Dinero:
            max_Dinero = Dinero
    return (contar, max_Dinero)
```

```
# Simulacion 
Soles = 50 # n soles
ganar = 0.3 #(ganar)
perder = 0.7 # (perder)
sims = [apostador(Soles, ganar,perder) for x in range(100)]
sims=np.array(sims)
```

```
sims.mean()
sims.shape
```

```
x, y = sims[:, 0:1], sims[:, 1:2]
plt.figure(figsize=(20,10))
plt.plot(x)
plt.title('Distribucion de intentos')
plt.show()
```

```
plt.figure(figsize=(20,10))
plt.plot(y)
plt.title('Distribucion de maximo dinero')
plt.show()
```
